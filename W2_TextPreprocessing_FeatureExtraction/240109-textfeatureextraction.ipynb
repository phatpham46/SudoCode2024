### **1. Load data**

```python
import pandas as pd
df = pd.read_csv('/kaggle/input/processed-data/processed_data.csv')
df.head()
```

```python
df.shape
```

```python
MAX_FEATURES = 10000
```

### **2. BOW Unigram**

```python
from sklearn.feature_extraction.text import CountVectorizer


# Initialize CountVectorizer for BOW
vectorizer_bow_unigram = CountVectorizer(max_features=MAX_FEATURES, analyzer='word', ngram_range=(1, 1)) 

# Fit and transform the tokenized content
X_bow = vectorizer_bow_unigram.fit_transform(df['tokenized_content'])

# Convert to DataFrame for better visualization
bow_df = pd.DataFrame(X_bow.toarray(), columns=vectorizer_bow_unigram.get_feature_names_out())
bow_df.head()
```

### **3. BOW Bigram**

```python
# Initialize CountVectorizer for BOW
vectorizer_bow_bigram = CountVectorizer(max_features=MAX_FEATURES, analyzer='word', ngram_range=(2, 2)) 

# Fit and transform the tokenized content
bow_bigram_features = vectorizer_bow_bigram.fit_transform(df['tokenized_title'])

# Convert to DataFrame for better visualization
bow_bigram_df = pd.DataFrame(bow_bigram_features.toarray(), columns=vectorizer_bow_bigram.get_feature_names_out())
bow_bigram_df.head()
```

### **4. TFIDF Unigram**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer_tfidf_unigram = TfidfVectorizer(max_features=MAX_FEATURES, analyzer='word', ngram_range=(1, 1))
tfidf_unigram_features = vectorizer_tfidf_unigram.fit_transform(df['tokenized_content'][:50000])

# Convert to DataFrame for better visualization
tfidf_unigram_df = pd.DataFrame(tfidf_unigram_features.toarray(), columns=vectorizer_tfidf_unigram.get_feature_names_out())
tfidf_unigram_df.head()
```

### **5. TFIDF Bigram**

```python
vectorizer_tfidf_bigram = TfidfVectorizer(max_features = MAX_FEATURES, analyzer='word', ngram_range=(2, 2))
tfidf_bigram_features = vectorizer_tfidf_bigram.fit_transform(df['tokenized_content'][:100000])

# Convert to DataFrame for better visualization
tfidf_bigram_df = pd.DataFrame(tfidf_bigram_features.toarray(), columns=vectorizer_tfidf_bigram.get_feature_names_out())
tfidf_bigram_df.head()
```
